```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
def gradient_descent(x,y):
    m, b, cost, iteration, n, rate, prev_cost = 0, 0, 0.00, 10000, len(x), 0.0002, 0
    for i in range(iteration):
        y_predicted = m * x + b
        cost = (1/n) * sum(pow(y - y_predicted, 2))
        md = (-2/n) * sum(x * (y - y_predicted))
        bd = (-2/n) * sum(y - y_predicted)
        m = m - rate * md
        b = m - rate * bd
        if math.isclose(cost, prev_cost, rel_tol=1e-20):
            break
        prev_cost = cost
        print(f'm: {m}, b: {b}, cost: {cost}, iteration: {i}')
```
