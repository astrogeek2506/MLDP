```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
```

```python
df = pd.read_csv('sonar_dataset.csv', header=None)
df
```

```python
X = df.drop(60, axis='columns')
Y = df[60]
Y.replace({'R' : 1, 'M' : 0}, inplace=True)

x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.9, random_state=10)
x_train.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(40, input_shape=(60,), activation='relu'),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid'),
])

model.compile(
    optimizer='adam',
    loss =  'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
model.fit(x_train, y_train, epochs=100, batch_size=10)
```

```python
model.evaluate(x_test, y_test)
```

```python
model_dropout = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dropout(0.5),
#     keras.layers.Dense(15,activation='relu'),
#     keras.layers.Dropout(0.5),                         
    keras.layers.Dense(1, activation='sigmoid'),
])

model_dropout.compile(
    optimizer='adam',
    loss =  'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
model_dropout.fit(x_train, y_train, epochs=100, batch_size=10)
```

```python
model_dropout.evaluate(x_test, y_test)
```

```python
y_pred = model_dropout.predict(x_test).reshape(-1)
y_pred = np.round(y_pred)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```
